```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
# Takes 30 minutes to run, you can encounter timeout error
# for this reason timeout is set to 10 to avoid this issue
data = pd.read_csv("C:\\Users\\Jakub\\Documents\\WEC2024_P-value_Abusers\\startup\\_data\\data_municipalities.csv")
geolocator = Nominatim(user_agent="my_geocoder")
names = data["municipality_name"]
long = []
lat =[]
for name in names:
    location = geolocator.geocode(name, timeout=10)
    lat.append(location.latitude)
    long.append(location.longitude)
```

```python
data["Longitude"] = long
data["Latitude"] = lat
data
```

```python
data.to_csv("spatial_data_munucipalities.csv", sep=",", index=False, header=True)
```
